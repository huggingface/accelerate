```python
from transformers import GPT2Tokenizer

# Carichiamo il tokenizer standard di GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Facciamo una prova: trasforma una frase in numeri (Token)
frase = "Il mio primo modello vuoto."
tokens = tokenizer.encode(frase)
print(f"Frase: {frase}")
print(f"Numeri (Token): {tokens}")
```
